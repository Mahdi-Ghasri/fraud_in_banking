````python
# TODO: بررسی هیستوگرام
plot_numeric_distributions(df)
````

````python
# TODO: شناسایی داده های پرت به کمک IQR
def find_outliers(df_copy):

    numeric_cols = df_copy.select_dtypes(include=['float64', 'int64']).columns

    for col in numeric_cols:
        Q1 = df_copy[col].quantile(0.25)
        Q3 = df_copy[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df_copy[(df_copy[col] < lower_bound) | (df_copy[col] > upper_bound)]

        print(f"{col}: {len(outliers)} outliers ({100 * len(outliers) / len(df_copy):.2f}%)")

        plt.figure(figsize=(6, 1.5))
        sns.boxplot(x=df_copy[col], color='skyblue')
        plt.title(f"Boxplot of {col}")
        plt.xlabel("")
        plt.tight_layout()
        plt.show()

find_outliers(df_copy)
````

````python
# TODO: تبدیل فرمت
def convert_timestamp(df, time_col='Timestamp'):
    df = df.copy()
    df[time_col] = pd.to_datetime(df[time_col],errors='coerce')
    df['year'] = df[time_col].dt.year
    df['month'] = df[time_col].dt.month
    df['day'] = df[time_col].dt.day
    df['hour'] = df[time_col].dt.hour
    df['minute'] = df[time_col].dt.minute
    df['dayofweek'] = df[time_col].dt.dayofweek
    df['is_weekend'] = df['dayofweek'].apply(lambda x: 1 if x >= 5 else 0)
    return df

df_copy=convert_timestamp(df_copy)
````

````python
def extract_time_features(df):
    """
    استخراج ویژگی‌های زمانی از ستون Timestamp و افزودن به دیتافریم.
    """
    # تبدیل Timestamp به datetime
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

    # استخراج ویژگی‌های زمانی
    df['Hour'] = df['Timestamp'].dt.hour
    df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
    df['Month'] = df['Timestamp'].dt.month
    df['Day'] = df['Timestamp'].dt.day

    # ویژگی‌های مشتق‌شده
    df['Is_Night'] = df['Hour'].apply(lambda x: 1 if 0 <= x <= 6 else 0)
    df['Is_Business_Hour'] = df['Hour'].apply(lambda x: 1 if 9 <= x <= 17 else 0)
    df['Is_Weekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)

    return df

````

````python
# TODO: متد زیر را کامل کنید.
def impute_missing_data(df, strategy_dict=None, auto_impute=True):
    """
    داده‌های گمشده را با استراتژی مشخص یا به صورت خودکار پر می‌کند.

    پارامترها:
        df: دیتافریم ورودی
        strategy_dict: دیکشنری از ستون‌ها و استراتژی‌ها (مثلاً {'age': 'mean', 'job': 'mode'})
        auto_impute: اگر True باشد، به صورت خودکار بر اساس نوع داده پر می‌کند

    خروجی:
        دیتافریم با داده‌های پر شده
    """
    df = df.copy()

    if auto_impute:
        for col in df.columns:
            if df[col].isnull().sum() == 0:
                continue
            if df[col].dtype in ['int64', 'float64']:
                df[col].fillna(df[col].mean(), inplace=True)
            else:
                df[col].fillna(df[col].mode().iloc[0], inplace=True)
    else:
        if strategy_dict is None:
            raise ValueError("strategy_dict باید وقتی auto_impute=False است، مشخص شود.")

        for col, strategy in strategy_dict.items():
            if col not in df.columns or df[col].isnull().sum() == 0:
                continue

            if strategy == 'mean':
                df[col].fillna(df[col].mean(), inplace=True)
            elif strategy == 'median':
                df[col].fillna(df[col].median(), inplace=True)
            elif strategy == 'mode':
                df[col].fillna(df[col].mode().iloc[0], inplace=True)
            elif strategy == 'unknown':
                df[col].fillna('unknown', inplace=True)
            elif strategy == 'knn':
                # فقط روی ستون عددی قابل اعمال است
                if df[col].dtype in ['float64', 'int64']:
                    imputer = KNNImputer(n_neighbors=3)
                    df[[col]] = imputer.fit_transform(df[[col]])
                else:
                    raise ValueError(f"استراتژی 'knn' فقط برای ستون‌های عددی قابل استفاده است: {col}")
            else:
                raise ValueError(f"استراتژی نامعتبر برای ستون {col}: {strategy}")

    return df
````

````python
# TODO: داده ها را استاندارسازی کنید.
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
X_train_numeric = X_train[numeric_cols]
X_test_numeric = X_test[numeric_cols]

# اعمال استانداردسازی
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_numeric)
X_test_scaled = scaler.transform(X_test_numeric)
````

````python
# TODO: متوازن سازی کلاس هدف

samplers = {
    "SMOTE": SMOTE(random_state=42),
    "SMOTETomek": SMOTETomek(random_state=42),
    "SMOTEENN": SMOTEENN(random_state=42),
    "ADASYN": ADASYN(random_state=42),
    "RandomUnderSampler": RandomUnderSampler(random_state=42)
}

best_sampler_name = None
best_f1 = 0
best_X_train_res = None
best_y_train_res = None

for name, sampler in samplers.items():
    X_res, y_res = sampler.fit_resample(X_train_scaled, y_train)
    model = XGBClassifier(tree_method='hist', use_label_encoder=False, eval_metric='logloss', random_state=42)
    model.fit(X_res, y_res)
    y_pred = model.predict(X_test_scaled)
    f1 = f1_score(y_test, y_pred)
    print(f"{name}: F1 Score = {f1:.4f}")
    if f1 > best_f1:
        best_f1 = f1
        best_sampler_name = name
        best_X_train_res, best_y_train_res = X_res, y_res

print(f"\nBest Sampler: {best_sampler_name} (F1 = {best_f1:.4f})")
````

````python
# TODO :  تعریف مدل‌هاو آموزش مدل ها
models = {
    "XGBoost": XGBClassifier(tree_method='hist', use_label_encoder=False, eval_metric='logloss', random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42),
    "LightGBM": lgb.LGBMClassifier(random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(best_X_train_res, best_y_train_res)
    y_pred = model.predict(X_test_scaled)
    f1 = f1_score(y_test, y_pred)
    results[name] = f1
    print(f"{name}: F1 Score = {f1:.4f}")
````
